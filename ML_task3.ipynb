```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from skimage.io import imread
from skimage.transform import resize
import os
```

```python
datadir='train3/'
images = []
labels = []
for i in os.listdir(datadir):
    img_path = os.path.join(datadir,i)
    img = imread(img_path)
    img_resized = resize(img, (64,64))
    images.append(img_resized.flatten())  # Flatten the image
    # labels.append(label)
    if 'cat' in i:
        labels.append(0)
    if 'dog' in i:
        labels.append(1)

labels = np.array(labels)
images = np.array(images)
```

```python
print(images)
```

```python
df=pd.DataFrame(images)
df['labels']=labels
df.head()
```

```python
y=df.labels
x=df.drop("labels",axis=1)
```

```python
y
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
svm = SVC(kernel='linear')  # You can experiment with other kernels like 'rbf'
svm.fit(x_train, y_train)
```

```python
y_pred = svm.predict(x_test)
print(f'Accuracy: {accuracy_score(y_test, y_pred)}')
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'C': [10, 20, 50],
    'kernel': ['poly', 'rbf']
}
GCV = GridSearchCV(SVC(kernel="poly"),param_grid,return_train_score=False)
GCV.fit(x_train,y_train)
```

```python
GCV.best_params_
```

```python
GCV.best_score_
```

```python
best_estimator=GCV.best_estimator_
y_pred = best_estimator.predict(x_test)
accuracy_score(y_test, y_pred)
```
